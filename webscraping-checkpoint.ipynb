```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?class=2;id=2019;type=year').text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup.prettify())
```

```python
match = soup.find('table', class_='engineTable').text
```

```python
match = match.split('\n\n\n')[2:-1]
```

```python
match[0] = match[0][2:]
```

```python
match[0].split('\n')
```

```python
matches = []

for game in match:
    matches.append(game.split('\n'))
```

```python
matches
```

```python
import pandas as pd
```

```python
matches_dataframe = pd.DataFrame(data=matches, columns=['Team 1', 'Team 2', 'Winner', 'Margin', 'Ground', 'Match Date', 'Scorecard'])
```
